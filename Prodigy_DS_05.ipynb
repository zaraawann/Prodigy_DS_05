```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('C:\\US_Accidents.csv')

# Data Exploration
print(df.info())
print(df.head())

# Data Cleaning
# Handle missing values, duplicates, and inconsistencies

# Exploratory Data Analysis (EDA)
# Example: Distribution of accidents by time of day
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour
plt.figure(figsize=(10, 6))
sns.histplot(df['Hour'], bins=24, kde=False)
plt.title('Distribution of Accidents by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()

# Example: Relationship between accidents and weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Condition', data=df, order=df['Weather_Condition'].value_counts().iloc[:10].index)
plt.title('Accidents by Weather Condition (Top 10)')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

# Visualization
# Example: Heatmap of accident hotspots
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Start_Lng', y='Start_Lat', data=df, alpha=0.01)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()


```
